### Purpose: Load the wine datasets (Red, White, Sparkinling, Rose) and clean the data so it can be loaded into SQL

```python
# Import dependencies
import pandas as pd
import os
import re
```

```python
# Load in the red wine data
red_data = os.path.join("Resources","Red.csv")
red_df = pd.read_csv(red_data)

red_df.head()
```

```python
# Add a column for wine type
red_df["Variety"] = 'Red'
red_df.head()
```

```python
# Load in the white wine data
white_data = os.path.join("Resources","White.csv")
white_df = pd.read_csv(red_data)

white_df.head()
```

```python
# Add a column for wine type
white_df["Variety"] = 'White'
white_df.head()
```

```python
# Load in the rose data
rose_data = os.path.join("Resources","Rose.csv")
rose_df = pd.read_csv(rose_data)

rose_df.head()
```

```python
# Add a column for wine type
rose_df["Variety"] = 'Rose'
rose_df.head()
```

```python
# Load in the rose data
sparkling_data = os.path.join("Resources","Sparkling.csv")
sparkling_df = pd.read_csv(sparkling_data)

sparkling_df.head()
```

```python
# Add a column for wine type
sparkling_df["Variety"] = 'Sparkling'
sparkling_df.head()
```

```python
compiled_df = pd.concat([red_df, white_df, rose_df, sparkling_df], axis=0)
compiled_df
```

```python
# Export the compiled dataframe to a csv
compiled_df.to_csv('Resources/compiled_wine.csv', index=False)
```
